```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
```

```python
from scipy.stats import uniform
size = 10000
start= 10
width = 20
data = uniform.rvs(size=size, loc = start, scale = width)
#print(data)
ax = sns.distplot(data,
             bins= 100,
             kde=True,
             color='orange',
             hist_kws={"linewidth":15, 'alpha': 1})
ax.set(xlabel='Uniform Distribution', ylabel='Frequency')
```

```python
from scipy.stats import norm
data = norm.rvs(size=100000, loc=100, scale=500)
print(data)
ax= sns.distplot(data,
                 bins = 100,
                 kde=True,
                 color='pink',
                 hist_kws={"linewidth": 15, 'alpha': 1})
ax.set(xlabel= 'Normal Distribition', ylabel = 'Frequency')
```
