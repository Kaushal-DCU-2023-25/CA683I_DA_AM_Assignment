### Statistical Testing

1. Null and Alternate Hypothesis

2. Set precision level

3. Gather sample data and calculate sample statistics

```python
import numpy as np
from scipy import stats

xa=30500
xb=30200
var_a = 9200**2
var_b = 8900**2
na=100
nb=100

s = np.sqrt((var_a + var_b)/2)
print(f'Standard deviation from var_a and var_b is : {s}')

t = ((xa - xb)/ np.sqrt((var_a/na) + (var_b/nb)))

print(f't-statistics is {t}')

df= na + nb -2

print(f'degre of freedom is {na + nb -2}')

p = 1 - stats.t.cdf(t, df=df)

print("t = " + str(t))
print("p = " + str(2*p))
```

```python
N=10

a = np.random.randn(N) + 2 # Normal/Gaussian distribution with mean 2, var 1
b = np.random.randn(N) # Normal/Gaussian distribution with mean 0, var 1

var_a = a.var(ddof=1)
var_b = b.var(ddof=1)

std_dev = np.sqrt((var_a + var_b)/2)
print(std_dev)

t = (a.mean() - b.mean()) / (std_dev * np.sqrt(2/N))
df = (2*N) - 2;

p = 1 - stats.t.cdf(t, df=df)

print(f't = { str(t)} and p = { str(2 * p)}')

t2, p2 = stats.ttest_ind(a, b)

print(f't2 = {t2} and p2 = {p2}')

```

```python
#Chi-square-test: 2 or more group...and targetus nominal variable
import numpy as np
a1 = [80, 220]
a2 = [50, 50]
a3 = [40, 10]
a4 = [30,20]

data_points = np.array([a1, a2, a3, a4])

chi2, p, dof, ex = stats.chi2_contingency(data_points)

print(f'chi2: {chi2}, p : {p}, dof : {dof}, Contingency Table i.e ex : {ex} ')
```

```python
# ANOVA: when comparing 2 or more variable for an outcome variable

import pandas as pd

df = pd.read_csv("https://www.reneshbedre.com/assets/posts/anova/onewayanova.txt", sep="\t")
print(df.describe)

#df.boxplot()

fvalue, pvalue = stats.f_oneway(df['A'], df['B'], df['C'], df['D'])
print(fvalue, pvalue)

#Get ANOVA table as R like output
import statsmodels.api as sm
from statsmodels.formula.api import ols
#reshape the d dataframe suitable for statsmodels package
d_melt = pd.melt(df.reset_index(), id_vars=['index'], value_vars=['A', 'B', 'C', 'D'])
print(d_melt)
d_melt.columns = ['index', 'treatments', 'value']
model = ols('value ~ C(treatments)', data=d_melt).fit()
anova_table = sm.stats.anova_lm(model, tpy=2)
print(anova_table)
```

```python
import scipy.stats as stats
w, pvalue = stats.levene(df['A'], df['B'], df['C'], df['D'])
print(w, pvalue)

# load packages
import scipy.stats as stats
w, pvalue = stats.shapiro(model.resid)
print(w, pvalue)
```
